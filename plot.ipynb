```python
#line plots
data = {
    'grade_6':['Patience', 'Joy', 'Ann', 'Lucy', 'Hannah', 'liz'],
    'Marks':[360, 400, 357,334,200,205],
    'weight':[5500, 6500, 5000, 5000, 4800,4900]
}
dataframe = pd.DataFrame(data)
plt.plot(dataframe['Marks'], dataframe['weight'])

#display the plot
plt.show()
```

```python
 #how to add grid lines
data = {
     'Cricket_bat':['SG', 'BOM', 'SS', 'KK', 'SP','ll'],
     'MRP': [2000, 2200, 2400, 2700, 2800, 3000],
     'Weight_grams':[1100, 1200,1250,1330,1480,1600] 
}
dataframe1 = pd.DataFrame(data)
plt.plot(dataframe1['MRP'], dataframe1['Weight_grams'])

#adding labels to a plot
plt.title('plot of bat price against weight', loc='center')
plt.xlabel("Price")
plt.ylabel('weight')
#add grid line and show plot
plt.grid()
plt.show()
```

```python
#add a matplotlib legend
a=np.arange(5)
b = [2,4,6,8,10]
c =[5,6,7,8,9]

fig =plt.figure()
ax =plt.subplot()

ax.plot(a,b ,'k--', label= 'frequency')
ax.plot(a,c,'k--',label= 'Periods')

#adding and positioning legend
legend = ax.legend(loc='lower right',fontsize='xx-small')

#set background color of a legend
legend.get_frame().set_facecolor('red')

#change font size of a legend
plt.title('frequency of a signal')
plt.show
```

```python
#bar graph in matplotlib
student = np.array(['Patience', 'Joy', 'Ann', 'Lucy', 'Hannah', 'liz'])
marks = np.array([360, 400, 357,334,200,205])
plt.bar(student, marks)

#adding label and title
plt.xlabel('student')
plt.ylabel('marks')
plt.title('student  marks')
plt.show()
```

```python
student = np.array(['Patience', 'Joy', 'Ann', 'Lucy', 'Hannah', 'liz'])
marks = np.array([360, 400, 357,334,200,205])
plt.pie(marks, labels=student,autopct='%1.2f%%')

#adding label and title
plt.xlabel('student')
plt.ylabel('marks')
plt.title('student  marks')
plt.show()
```

```python
student = np.array(['Patience', 'Joy', 'Ann', 'Lucy', 'Hannah', 'liz'])
marks = np.array([360, 400, 357,334,200,205])
plt.hist(marks, bins=12)

#adding label and title
plt.xlabel('student')
plt.ylabel('marks')
plt.title('student  marks')
plt.show()
```

```python
#plot a histogram
arr= np.array([10,50,34,67,21,7,59,62,45,48,10,8,41,32,66,59,18,79,77,65])
plt.hist(arr, bins=[0,20,40,60,80,100])
plt.xlabel('marks')
plt.ylabel('student')
plt.title('student marks')
plt.show()
```

```python
#scatter plot
team1_score =[25,47,34,38,27,40,42,18]
team2_score =[7,22,40,29,27,10,19,31]

scoreRange = [5,10,15,20,25,30,35,40]
plt.scatter(team1_score, scoreRange, color ='r')
plt.scatter(team2_score, scoreRange, color ='b')
plt.xlabel('team score')
plt.ylabel('score range')
plt.title('score of two teams')
```
