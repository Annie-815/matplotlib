```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = np.linspace(0, 10, 100)
y = np.sin(np.pi * X/2) + 2
y2 = 10 -(X-5)**2
plt.plot(X,y)
plt.show()
```

```python
#drawing a horizontal line with pyplot
plt.plot(X, y, zorder=2, label='sin wave')
plt.axhline(2, color ='red', lw=4, zorder=1, label='baseline')
plt.legend()  
```

```python
#adding a vertical line
plt.plot(X, y2, label='sin wave')
plt.axvline(5,label='baseline',color='purple', ls='--')
plt.legend(loc='upper left')
```

```python
#Adding Symbols with tex
x = np.linspace(1,4,1000)
y =x
y2 = x**2
y_log = np.log(x)
y_cos = np.cos(np.pi *x)
```

```python
#add superscript
plt.plot(x,y2)
plt.title(r'$y=x^2$')
plt.show()
```

```python
#add subscript
plt.plot(x,y)
plt.title(r'$y=x_i$')
plt.show()
```

```python
#add letters and math functions
plt.plot(x,y_cos)
plt.text(2.3,0.75, r'$y=\cos(\pi x)$')
plt.show()
```

```python
plt.plot(x,y_cos)
plt.title(r'$y=\cos(\pi x)$')
plt.text(2.3, 0.75, r'$y=\cos(\pi x)$')
plt.xlabel(r'$y=\cos(\pi x)$')
plt.ylabel(r'$y=\cos(\pi x)$')
plt.legend([r'$y=\cos(\pi x)$'],loc=4)
plt.show()
```

```python
#adding fractions
x= np.linspace(0.5,100,1000)
y_frac = 2+ 1/(x**2)
```

```python
plt.plot(x,y_frac)
plt.ylim(0, None)
plt.title(r'$y=2+\frac{1}{x^2}$',fontsize=20)
plt.text(40,4,r'$\lim_{x\rightarrow\infty}\;2 + \frac{1}{x^2} =2 $', fontsize=20)
```
